```python
from collections import Counter

def the_longest_and_most_common_word(text):
    lst_no = ['.', ',', ':', '!', '"', "'", '[', ']', '-', '—', '(', ')'  ]   # и т.д.
    lst = []

    for word in text.lower().split():
        if not word in lst_no:
            wordnew = word  
            lst.append(wordnew)
    longest_word = max(lst, key=len)
    print(longest_word)
    

    c = Counter(lst)
    c=c.most_common(1)[0]
    print(c[0])
    
    return longest_word, c[0]
```

```python
import pandas as pd

OurDataFrame = {
    'FileName' : [],
    'LongestWord' : [],
    'MostCommonWord' : [],
    'Mark_Label':[]
}
```

```python
# data loading

import os
train_dir = 'aclImdb/train/'

labels = []
texts = []

# Стараемся думаем
# Стараемся думаем
# 1.Перебрать 2 папки
# 2. перебрать все файлы
#  3. считать файл,добавить в texts
#  4. метки: если имя папки == neg,-> labels.append(0) else labels.append(1)
#
#
#

for label_type in ['neg', 'pos']:
    
    dir_name = train_dir+label_type+'/'
    
    for fname in os.listdir(dir_name):
        
        if fname[-4:]=='.txt':
            
            OurDataFrame['FileName'].append(fname)
            
            f = open(os.path.join(dir_name, fname), encoding='utf-8')
            #f = open(dir_name + fname, encoding='utf-8')
            text = f.read()
            f.close()
                
            #the_longest_and_most_common_word(text)    
            longest_word, c = the_longest_and_most_common_word(text)
            
            OurDataFrame['LongestWord'].append(longest_word)
            OurDataFrame['MostCommonWord'].append(c)
            
            
            
            if label_type == 'neg':
                OurDataFrame['Mark_Label'].append(0)
            else:
                OurDataFrame['Mark_Label'].append(1)
```
