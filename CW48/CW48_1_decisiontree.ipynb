```python
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np
import pipes_2 as pp
from sklearn.tree import DecisionTreeClassifier     ### замена класса,меняем на другое дальше
```

```python
df = pd.read_csv('air_train.csv')
df = df.drop(['Unnamed: 0','id'], axis = 1)
df = pp.myPipe.transform(df)
X = df.drop(['satisfaction'], axis = 1)
y = df.satisfaction

t = pd.read_csv('air_test.csv')
t = t.drop(['Unnamed: 0','id'], axis = 1)
t = pp.myPipe.transform(t)
X_t = t.drop(['satisfaction'], axis = 1)
y_t = t.satisfaction
```

```python
tree = DecisionTreeClassifier()    ### замена класса
```

```python
paramgrid = { 'max_depth' :[5,8,13,17,20], 'min_samples_split':[1,2,4,7,11], 'max_features':[3,4,6,8,15]}   ## замена параметров
```

```python
GS_cv = GridSearchCV(tree,param_grid = paramgrid, cv = 3, scoring = 'f1')
# cv - количество разбиений
```

```python
GS_cv.fit(X,y)
```

```python
GS_cv.best_params_
```

```python
best_gs_tree=GS_cv.best_estimator_
```

```python
best_gs_tree.score(X_t,y_t)
```

```python
best_gs_tree.score(X,y)
```

```python
from sklearn.metrics import f1_score
pred = best_gs_tree.predict(X_t)
f1_score = (y_t,pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cs = cross_val_score(best_gs_tree,X,y, cv= 5, scoring = 'f1')
```
