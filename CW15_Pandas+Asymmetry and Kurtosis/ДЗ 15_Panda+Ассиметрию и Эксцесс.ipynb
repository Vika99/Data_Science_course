```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
# читаем файл
df = pd.read_csv('Task_one.csv')
```

```python
# рассчитайте показатели вариации :
# 1.размах вариации
max1 = df['v1'].max()
min1 = df['v1'].min()
R1 = max1 - min1
max2 = df['v2'].max()
min2 = df['v2'].min()
R2 = max2 - min2

print("Размах вариации v1 = ", R1)
print("Размах вариации v2 = ",R2)
```

```python
# 2.среднее по выборке:

average1 = df['v1'].mean()  
# average1 = df['v1'].sum()/1000
average2 = df['v2'].mean()
# average2 = df['v2'].sum()/1000
print('Среднее по выборке v1 = ',average1)
print('Среднее по выборке v2 = ',average2)
```

```python
# 2.Среднее линейное отклонение (ВАРИАНТ 1):
# a. от наших значений в столбиках v1,v2 отнимаем среднее значение по выборке(смотри выше average1,2):
lin_div1 = df['v1'] - average1
lin_div2 = df['v2'] - average2

# print(lin_div1)
# print(lin_div2)

# b.Берем lin_div по модулю, суммируем значения и делим на общее количество и получаем наше стандартное отклонение.

standartDeviation1 = np.abs(lin_div1).sum()/1000
standartDeviation2 = np.abs(lin_div2).sum()/1000

print("стандартное отклонение по v1 = ",standartDeviation1)
print("стандартное отклонение по v2 = ",standartDeviation2)
```

```python
# 2.Среднее линейное отклонение (ВАРИАНТ 2):

stanDev1 = df['v1'].std()
stanDev2 = df['v2'].std()
print("стандартное отклонение по v1 = ",stanDev1)
print("стандартное отклонение по v2 = ",stanDev2)
```

```python
# 3. Мода — значение во множестве наблюдений, которое встречается наиболее часто.
mode1 = df['v1'].mode()
mode2 = df['v2'].mode()

print('Moda по v1 = ', mode1)
print('Moda по v2 = ', mode2)
```

```python
# 3. Медиана
median1 = df['v1'].median()
median2 = df['v2'].median()

print('Median по v1 = ', median1)
print('Median по v2 = ', median2)
```

```python
# 4. Ассиметрия (skew)
skew1 = df['v1'].skew()
skew2 = df['v2'].skew()

print('Ассиметрия по v1 = ', skew1)
print('Ассиметрия по v2 = ', skew2)
```

```python
# 4. Эксцесс (kurt)

kurt1 = df['v1'].kurt()
kurt2 = df['v2'].kurt()

print('Эксцесс по v1 = ', kurt1)
print('Эксцесс по v2 = ', kurt2)
```

```python
print("Размах вариации v1 = ", R1)
print("Размах вариации v2 = ",R2)

print('Среднее по выборке v1 = ',average1)
print('Среднее по выборке v2 = ',average2)

print("стандартное отклонение по v1 = ",stanDev1)
print("стандартное отклонение по v2 = ",stanDev2)

print('Median по v1 = ', median1)
print('Median по v2 = ', median2)

print('Ассиметрия по v1 = ', skew1)
print('Ассиметрия по v2 = ', skew2)

print('Эксцесс по v1 = ', kurt1)
print('Эксцесс по v2 = ', kurt2)
```
