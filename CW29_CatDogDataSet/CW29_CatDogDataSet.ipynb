## разделение данных на тестовое и обучающее множество

```python
import os, shutil
import random as rn
```

```python
train_dirs = ['PetImages/train/Dog', 
              'PetImages/train/Cat']

test_dirs =  ['PetImages/validation/Dog', 
              'PetImages/validation/Cat']
```

```python
train_dirs = ['PetImages/train/Dog', 
              'PetImages/train/Cat']

test_dirs =  ['PetImages/test/Dog', 
              'PetImages/test/Cat']
```

```python
# Страшный код
for i in range(len(train_dirs)):
    fnames = os.listdir(train_dirs[i])
    toCopy = len(fnames)//3

    filesToCopy = []
    for j in range(toCopy):
        filesToCopy.append(rn.choice(fnames))
    
    # cOPY
         
    for fi in filesToCopy:
        src = os.path.join(train_dirs[i],fi)
        dst = os.path.join(test_dirs[i], fi)
        shutil.copyfile(src,dst)

    
    for fi in filesToCopy:
        try:
            src = os.path.join(train_dirs[i],fi)
            os.remove(src)
            print(src+' deleted')
            #time.sleep(1)
        except:
            pass
```

```python
#фиксируем размер картинок
Pic_Y, Pic_X = 180, 180
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#создаем объект  train_datagen, где rescale у нас означает,что все входные данные будут делиться на 255.
# мы таким образом загоняем наши данные в интервал от 0 до 1.
train_datagen = ImageDataGenerator(rescale = 1./255)
# train_generator означает, что сделать объект нам из директории 'PetImages/train, 
#   target_size = (Pic_Y, Pic_X) - означает извлекать размером высота на ширину
#
# class_mode = 'binary' - тип предоставления картинок. Если бы было больше двух классов(у нас собака и кот) использовали бы другое.
#                            
train_generator = train_datagen.flow_from_directory('PetImages/train',
                                                   target_size = (Pic_Y, Pic_X),
                                                   batch_size = 64,
                                                   class_mode = 'binary')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
                            
test_generator = test_datagen.flow_from_directory('PetImages/validation',
                                                   target_size = (Pic_Y, Pic_X),
                                                   batch_size = 64,
                                                   class_mode = 'binary')
```

```python
# Обучение модели с 0

model = models.Sequential()

model.add(layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape = (Pic_Y, Pic_X, 3)))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=64, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=128, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=256, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Conv2D(filters=256, kernel_size=3, activation="relu"))

model.add(layers.Flatten())
model.add(layers.Dense(1, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
# Новый интересный объект
callbacks = [tf.keras.callbacks.ModelCheckpoint(
        filepath="convnet_from_scratch.h5",
        save_best_only=True,
        monitor="val_loss")]
```

```python
history = model.fit(train_generator, epochs=5, validation_data=test_generator,
    callbacks=callbacks)
```

## Перенос знаний!

```python
# закгрузка сверточных слоев VGG16

conv_base = tf.keras.applications.vgg16.VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(180, 180, 3))
```

```python
# Заморозка сверточных слоев
conv_base.trainable = False
```

```python
# Конструируем модель
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
callbacks = [tf.keras.callbacks.ModelCheckpoint(
        filepath="VGG_FIX_model.h5",
        save_best_only=True,
        monitor="val_loss")]
```

```python
# фитим на тех же генераторах то есть с расширением данных
history = model.fit(train_generator, epochs=1, validation_data=test_generator, 
    callbacks=callbacks)
```
