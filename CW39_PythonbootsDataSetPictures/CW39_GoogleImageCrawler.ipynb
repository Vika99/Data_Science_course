```python
for keyw in ['Adidas','Nike', 'converse']:
    gc = GoogleImageCrawler(parser_threads = 2,
                           downloader_threads = 4,
                           storage = {'root_dir': 'IMAGES/{}'.format(keyw)})
    gc.crawl(keyword = keyw, max_num = 50, min_size = (200,200))
```
