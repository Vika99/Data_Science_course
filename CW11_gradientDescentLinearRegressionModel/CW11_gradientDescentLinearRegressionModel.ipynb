```python
def draw_f (x0, x1, *ff):
    
    for f in ff:
        x = np.linspace(x0,x1,100)
        y = np.linspace(x0,x1,100)
    
        for i in range(x.shape[0]):
            y[i] = f(x[i])
    
        plt.plot(x,y)
    
    plt.show()
```

```python
#градиентный спуск
```

```python
# генерируем данные
# X формируем матрицу случайных чисел с равномерным распределением,двумерная матрица,100 строк и 1 столбец
X = 2*np.random.rand(100,1)
# y тоже матрица,и мы добавляем 4+3*X ,это будет близко к нашей прямой
y = 4+3*X + np.random.randn(100,1) # шум
```

```python
plt.scatter(X,y)
```

```python
# теперь задача построить линейную регрессию методом градиентного спуска
# мы добавялем столбец единиц,для того,что,когда будет осуществлять вектроное умножение, у нас есть свободный член 4 (в y)
# и нам его тоже надо искать и надо добавить единицы.с помощью функции np.c,она добавляет столбцы
#  это значит,что мы добавляем впереди столбцы единиц и потом нашу матрицу X
# X_b = np.c_[np.ones((100,1)),X]  
```

```python
# это значит,что мы добавляем впереди столбцы единиц и потом нашу матрицу X
X_b = np.c_[np.ones((100,1)),X]  
```

```python
X_b
```

```python
# lr наш параметр, скорость обучения, на него будем сдвигаться,на каждом шаге считаем градиент и сдвигаемся на какое-то небольшое число
lr = 0.1 # скорость обучения
# iters : количество итераций,сколько раз гоняем наш алгоритм
# iters = 100
iters = 10
# m: количество веторов
m = 100
# h: наш искомый вектор модели,мы его случаным образом заполняем
h = np.random.randn(2,1) # инициализация




```

```python
h
```

```python
# поиск
for it in range(iters):
# мы считаем градиент по формуле матричной из лекции:
    gradient = (2/m)*X_b.T @ (X_b @ h - y)
# находим наш h, hкоторый был отнимаем градиент умноженный на скорость обучения
    h = h - lr * gradient
```
