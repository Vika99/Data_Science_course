```python
def f(x):
    return x*x
```

```python
def draw_f (x0, x1, *ff):
    
    for f in ff:
        x = np.linspace(x0,x1,100)
        y = np.linspace(x0,x1,100)
    
        for i in range(x.shape[0]):
            y[i] = f(x[i])
    
        plt.plot(x,y)
    
    plt.show()
```

```python
draw_f( -5, 2, lambda x: x**3+4*x**2, lambda x: 3*x**2+8*x, lambda x:0)
```

```python
draw_f( -5, 5, math.sin, math.cos)
```
