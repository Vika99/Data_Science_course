```python
from matplotlib import pyplot as plt
import math
import numpy as np
data = np.loadtxt('linreg_4.txt')
```

```python
# Отделите независимые переменные от целевой переменной
x,y=np.hsplit(data,[3])
```

```python
# адаптируйте написанный на уроке код градиентного спуска для нахождения коэффициентов линейной модели вида
# y = h0 + h1x1 + h2x2 + h3x3
# генерируем данные
# y = h0 + h1x1 + h2x2 + h3x3
# так как у нас  h0, то нам надо единичные столбцы
x_ = np.c_[np.ones((100,1)),x] 
```

```python
x_
```

```python
# lr наш параметр, скорость обучения, на него будем сдвигаться,на каждом шаге считаем градиент и сдвигаемся на какое-то небольшое число
lr = 0.1 # скорость обучения
# lr = 0.001
# iters : количество итераций,сколько раз гоняем наш алгоритм
# iters = 100
iters = 100
# m: количество веторов
m = 100
# h: наш искомый вектор модели,мы его случаным образом заполняем
h = np.random.randn(4,1) # инициализация
```

```python
 h
```

```python
# поиск
for it in range(iters):
# мы считаем градиент по формуле матричной из лекции:
    gradient = (2/m)*x_.T @ (x_ @ h - y)
# находим наш h, hкоторый был отнимаем градиент умноженный на скорость обучения
    h = h - lr * gradient
```

```python
h
```

```python
# поиск лучших векторов 
h_best = np.linalg.inv(x_.T.dot(x_)).dot(x_.T).dot(y)
```
