```python
f = open ('new_ratings.csv', 'r',encoding='utf-8')
```

```python
lines.pop(0)
```

```python
mat = []
for l in lines:
    newl = l.replace('\n', '').split(';')
    mat.append(newl)
```

```python
#1.	Общее количество оценок в файле.
marks = []
for i in mat:
    marks.append(i[2])
print(len(marks))   

```

```python
#2.	Общее количество пользователей, поставивших оценки
users = []
for l in mat:
    if l[0] not in users:
        users.append(l[0])
print("total numbet of users:", users)

print (len(users))
```
